```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
comp_list = list(set(df['종목코드'].values))
```

```python
final_open_df = pd.DataFrame()
final_high_df = pd.DataFrame()
final_low_df = pd.DataFrame()
final_close_df = pd.DataFrame()
final_volume_df = pd.DataFrame()

for i in range(0, len(comp_list)):
    data = fdr.DataReader(comp_list[i].replace('A', ''), '2020-10-31', '2023-05-30')

    open_df = pd.DataFrame(data['Open'])
    open_df.columns = [comp_list[i]]
    final_open_df = pd.concat([final_open_df, open_df], axis = 1)

    high_df = pd.DataFrame(data['High'])
    high_df.columns = [comp_list[i]]
    final_high_df = pd.concat([final_high_df, high_df], axis = 1)

    low_df = pd.DataFrame(data['Low'])
    low_df.columns = [comp_list[i]]
    final_low_df = pd.concat([final_low_df, low_df], axis = 1)

    close_df = pd.DataFrame(data['Close'])
    close_df.columns = [comp_list[i]]
    final_close_df = pd.concat([final_close_df, close_df], axis = 1)

    volume_df = pd.DataFrame(data['Volume'])
    volume_df.columns = [comp_list[i]]
    final_volume_df = pd.concat([final_volume_df, volume_df], axis = 1)
```

```python
final_open_df.to_csv('final_open_data.csv')
final_high_df.to_csv('final_high_data.csv')
final_low_df.to_csv('final_low_data.csv')
final_close_df.to_csv('final_close_data.csv')
final_volume_df.to_csv('final_volume_data.csv')
```

```python
final_close_df = final_close_df[1:]
```

```python
# 로그 수익률
final_ret = pd.DataFrame(np.log(final_close_df.iloc[-1]) - np.log(final_close_df.iloc[1]))
final_ret.columns = ['final_log_return']
final_ret = final_ret.sort_values('final_log_return', ascending = False)


```

```python
final_ret
```

```python
# 최종 매수 수익률 List
num = 200
long = final_ret[:num]
short = final_ret[-num:]

```

```python
# 총자산 최종 수익률
eafr = (long.sum() - short.sum()) / 400
print(eafr)
```

```python
# 연율화된 총자산 최종 수익률
aeafr = eafr * 250 / (len(final_close_df))

print(aeafr)
```

```python
# 전체 수익률
close_ret = np.log(final_close_df[1:]) - np.log(final_close_df.shift(1).dropna())
```

```python
# 선택된 200개 종목 리스트
long_list = list(long.index)
short_list = list(short.index)
```

```python
annualized_long_200 = close_ret[long_list].sum(axis = 1) * 250
annualized_short_200 = -close_ret[short_list].sum(axis = 1) * 250
```

```python
total_annualized = annualized_long_200 + annualized_short_200
```

```python
# 총자산 일간 수익률 변동성
vol_fin = np.sqrt(np.square(total_annualized - total_annualized.mean()).sum() / (len(final_close_df) - 2))

print(vol_fin)
```

```python
total_annualized - total_annualized.mean()
```

```python
# Sharpe index

sharpe = (aeafr - 0.035) / vol_fin
print(sharpe)
```
